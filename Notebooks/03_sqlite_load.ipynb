# 03 - SQLite Load

This notebook loads the processed UK electricity demand dataset into a local SQLite database.

The workflow includes:

- Loading the processed CSV file
- Creating a local SQLite database
- Writing the dataset to a database table
- Running basic SQL checks

```python
%reset -f

import pandas as pd
import sqlite3
from pathlib import Path
```

## Load Processed Dataset

```python
processed_file = Path("../Data/Processed/combined_energy_demand.csv")
df = pd.read_csv(processed_file)
df.head()
```

## Create SQLite Database Connection

```python
database_path = Path("../Data/Processed/energy_pipeline.db")
connection = sqlite3.connect(database_path)

print("SQLite database connection created successfully.")
```

## Load Dataset into SQLite Table

```python
df.to_sql(
    "energy_demand",
    connection,
    if_exists="replace",
    index=False
)

print("Dataset loaded into SQLite successfully.")
```

## Validate SQLite Table Load

```python
query = """
SELECT COUNT(*) AS total_rows
FROM energy_demand
"""

row_count = pd.read_sql_query(query, connection)
row_count
```

## Example SQL Aggregation Query

```python
query = """
SELECT
    SUBSTR(SETTLEMENT_DATE, 1, 4) AS year,
    ROUND(AVG(ENGLAND_WALES_DEMAND), 2) AS average_demand
FROM energy_demand
GROUP BY year
ORDER BY year
"""

average_demand = pd.read_sql_query(query, connection)
average_demand
```

## SQLite Load Summary

The processed UK electricity demand dataset was successfully loaded into a local SQLite database.

Validation confirmed:

- Successful database connection
- Successful table creation
- Correct row count after loading
- Successful execution of SQL aggregation queries

The dataset is now available for downstream querying and cloud upload operations.
