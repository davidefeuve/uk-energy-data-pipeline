# 04 - BigQuery Upload

This notebook uploads the processed UK electricity demand dataset to Google BigQuery.

The workflow includes:

- Loading the processed dataset
- Authenticating with Google Cloud
- Connecting to BigQuery
- Uploading the dataset to a BigQuery table
- Validating the cloud upload

```python
%reset -f

import pandas as pd
from pathlib import Path
from google.cloud import bigquery
```

## Authenticate with Google Cloud

```python
credentials_path = Path("../Credentials/chrome-energy-496120-j4-299aaa8130ba.json")
client = bigquery.Client.from_service_account_json(credentials_path)

print("BigQuery client created successfully.")
```

## Load Processed Dataset

```python
processed_file = Path("../Data/Processed/combined_energy_demand.csv")
df = pd.read_csv(processed_file)
df.head()
```

## Configure BigQuery Table

```python
table_id = ("chrome-energy-496120-j4.energy_data.energy_demand")
job_config = bigquery.LoadJobConfig(
    write_disposition="WRITE_TRUNCATE"
)

print("BigQuery table configuration created.")
```

## Upload Dataset to BigQuery

```python
load_job = client.load_table_from_dataframe(
    df,
    table_id,
    job_config=job_config
)

load_job.result()
print("Dataset uploaded to BigQuery successfully.")
```

## Validate BigQuery Upload

```python
query = """
SELECT COUNT(*) AS total_rows
FROM `chrome-energy-496120-j4.energy_data.energy_demand`
"""

query_result = client.query(query).to_dataframe()
query_result
```
